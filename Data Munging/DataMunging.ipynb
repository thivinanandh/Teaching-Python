<h2 style="color:green; font-family: 'Nunito Sans', sans-serif;">Regex</h2>

<div style="color:black; font-family: 'Quicksand', sans-serif;font-weight:500">


Regex stands for Regular Expression. it is one of the most powerfull pattern based matching tool which could be usefull in extracting most complex patterns present in the string. 

- [abc]  - single character of a or b or c
- [^abc] - a single character except a b or c
- [a-z]  - a single charecter in the range a to z
- .      - any single character
- /s     - any whitespace character
- /S     - any non whitespace character
- /d     - any digit
- /D     - any non Digit
- /w     - any word
- /W     - any nonWord character
- a?     - zero or one of appearance of character 'a'
- a*     - zero or more of a appearance of character 'a'
- a{3}   - exactly 3 of character a
- ^      - start of string
- $      - end of string

</div>

````python
SpamEmail = ["cmmail.com","nigerianprince.com","bounce.ml01.com","catchaplane.net","aoll.com"]

import pandas as pd

##read the data into a dataframe

emailDf  =  pd.read_csv("hotel_bookings.csv",delimiter=",")

email = emailDf["email_address"].copy(deep=True);
emailComment = [];

for mail in email:
    k = re.findall(r'\S+@\S+\.\S+$',mail)
    comment = "Not valid"
    if(len(k)):
        last = k[0].split('@')[1]
        if(last in SpamEmail):
            comment = "spam"
        else:
            comment = "valid"
    emailComment.append(comment)

emailDf["email_comment"] = emailComment
````

<div style="color:black; font-family: 'Quicksand', sans-serif;font-weight:500">


The mail data contains the mailing information from a sample server. it has various information like sender email, subject, Spam probability etc. 

we will try to use regex to extract those data to perform analysis on the same.
    
</div>

<code>

X-DSPAM-Probability: 0.0000

</code>

````python
import re
import sqlite3

update = False;
email = ""
prob  = 0.0;


emailList = [];
probList = [];

with open("mail.txt" , "r") as f:
    for line in f:
        if (not line.startswith("From"))  and (not line.startswith("X-DSPAM-Probability")) : continue;
        
        if(line.startswith("From")): 
            k = re.findall(r'^From (\S+@\S+\.\S+)',line)
            if(len(k) ==0 ): 
                continue;
            email = k[0]
            emailList.append(email);
        
        else:    ### Probability 
            k = re.findall(r'\S+: ([0-9.]+)',line)
            prob = float(k[0])
            probList.append(prob);
````

````python
df = pd.DataFrame(list(zip(emailList, probList)),
               columns =['email', 'spamProb'])
````

<h3 style="color:green; font-family: 'Nunito Sans', sans-serif;">Awk </h3>

<div style="color:black; font-family: 'Quicksand', sans-serif;font-weight:500">

Awk is well known for its ability to parse/tokenize data in column wise for data types like csv,tsv,ssv etc.

awk is mostly used with pipelines in the unix enviroiinment to get only the required data values. 

```cat server.log | awk '{if ($9 == 200) print $1}``` <br />
```cat server.log | awk '/.*/s200 {print $1}' | uniq -c | sort -nk1,1 | tail -n10```
    
</div>



<h3 style="color:green; font-family: 'Nunito Sans', sans-serif;">sed</h3>

<div style="color:black; font-family: 'Quicksand', sans-serif;font-weight:500">
    
sed is a stream editor. A stream editor is used to perform basic text transformations on an input stream (a file or input from a pipeline)

sed can use the regular expressions to identify patterns and replace them with custom strings
<br />
    
  

``` cat server.log | sed -E 's/^([0-9.]+).*\s([0-9]{3})\s[0-9].*$/\1 \2/' | tail -n10```
    
</div>
